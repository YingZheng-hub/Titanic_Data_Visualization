# 4.Bar Chart for Categorical Features
* Pclass
* Sex
* SibSp ( # of siblings and spouse)
* Parch ( # of parents and children)
* Embarked
* Cabin

```python
def bar_chart(feature):
    survived = train[train['Survived']==1][feature].value_counts()
    dead = train[train['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked= False, figsize=(10,5))
```

```python
bar_chart('Sex')
```

This chart confirms that male passengers are more likely dead in the accident. 

```python
bar_chart('Pclass')
```

* The chart confirms that Class 1 passengers are more likely to survive than passengers in Class 2 and 3.
* The chart also confirms that Class 3 passengers are more likely dead than ones in other classes. 


```python
bar_chart('SibSp')
```

* The charts confirms that passengers with more than 2 siblings or spouse are more likely survived. 
* The charts confirms that passengers with 0 siblings or spouse are more likely dead. 

```python
bar_chart('Parch')
```

* The chart confirms that passengers aboarded with two parents or children more likely survived.
* The chart also confirms that passengers aboarded with no parents or children more likely dead.

```python
bar_chart('Embarked')
```

* The chart confirms that passengers aboraded from S and Q more likely dead.
* The chart also confirms that passengers aboraded from C more likely survived.

# **5.2 Passenger's Title

# 6 Modelling 

```python
# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import numpy as np
```

# **6.2 Cross Validation (K-fold)****

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

# **6.2.1 kNN**

```python
clf = KNeighborsClassifier(n_neighbors = 13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# kNN Score
round(np.mean(score)*100, 2)
```

# 6.2.2 Decision Tree

```python
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# decision tree Score
round(np.mean(score)*100, 2)
```

# 6.2.3 Ramdom Forest

```python
clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python

# Random Forest Score
round(np.mean(score)*100, 2)
```

# 6.2.4 Naive Bayes

```python
clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Naive Bayes Score
round(np.mean(score)*100, 2)
```

# 6.2.5 SVM

```python
clf = SVC()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

# 7. Testing

```python

clf = SVC()
clf.fit(train_data, target)

test_data = test.drop("PassengerId", axis=1).copy()
prediction = clf.predict(test_data)
```
